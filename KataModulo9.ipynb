## Ejercicio1: Uso de funciones en Python

```python
def inform_combust(tanque1, tanque2, tanque3):
    total_average = ((tanque1 + tanque2 + tanque3) / 3)

    return f"""Informe Combustible:
    Total Promedio: {total_average}%
    Tanque Interno: {tanque1}%
    Tanque Externo: {tanque2}%
    Tanque Hidrogeno: {tanque3}% 
    """
print(inform_combust(55,65,79))


def promedio(values):
    total = sum(values)
    number_of_items = len(values)
    return total / number_of_items

promedio([80, 85, 81]) 


def inform_combust(tanque1, tanque2, tanque3):

    return f"""Informe Combustible:
    Total Promedio: {promedio([tanque1,tanque2,tanque3])}%
    Tanque Interno: {tanque1}%
    Tanque Externo: {tanque2}%
    Tanque Hidrogeno: {tanque3}% 
    """
print(inform_combust(55,65,79))
```
