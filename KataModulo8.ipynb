## Ejercicio2: Programaci√≥n dinamica con diccionarios

```python
planet_moons = {
    'mercury': 0,
    'venus': 0,
    'earth': 1,
    'mars': 2,
    'jupiter': 79,
    'saturn': 82,
    'uranus': 27,
    'neptune': 14,
    'pluto': 5,
    'haumea': 2,
    'makemake': 1,
    'eris': 1
}

moons = 0

for value in planet_moons.values():

    moons= moons + value

planets = len(planet_moons.keys())

promedio = moons / planets

print(f'{planets} planetas con un total de {moons} lunas y {round(promedio)} lunas en promedio')


```
