## Ejercicio2: Trabajando con datos de una lista

```python
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Neptune']
userplanets = input('Escribe el nombre de un planeta')

search_index = planets.index(userplanets)

print('Planeta m치s cerca de '+ userplanets)
print(planets[0:search_index])

print('Planeta m치s lejano de '+ userplanets)
print(planets[search_index + 1:])
```
