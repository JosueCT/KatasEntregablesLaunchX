## Trabajando con ELSE

```python
a = 23
b = 97
if a >= b:
    print(a)
else:
    print(b)
```

## Trabajando con ELSE IF(elif)

```python
a = 93
b = 27
if a >= b:
    print("a es mayor o igual que b")
elif a == b:
    print("a es igual que b")
```

## Trabajando con IF, ELSE y ELIF

```python
a = 93
b = 27
if a > b:
    print("a es mayor que b")
elif a < b:
    print ("a es menor que b")
else:
    print("a es igual que b")
```

## Trabajando con LCA(Logica Condicional Anidada)

```python
a = 26
b = 28
c = 27
if a > b:
    if b > c:
        print("a es mayor que b y b es mayor que c")
    else:
        print("a es mayor que b y menor que c")
elif a == b:
    print("a es igual que b")
else:
    print("a es menor que b")
```
