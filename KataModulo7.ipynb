## Ejercicio1: Bucle "while"

```python
new_planet = ''
planets = []

while new_planet.lower() != 'done':
    if new_planet:
        planets.append(new_planet)
    new_planet = input('Escribe un planeta para completar la lista:')

print(planets)
```

## Ejercicio2: Ciclo FOR

```python
for planet in planets:
    print(planet)
```
