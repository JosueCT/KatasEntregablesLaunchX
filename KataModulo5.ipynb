## Ejercicio 1: Utilizar Operadores aritméticos

```python
tierra = int('149597870')
jupiter = int('778547200')

distanciakm = jupiter - tierra
print(distanciakm)

distanciamillas = distanciakm * 0.621
print(round(distanciamillas))
```

## Ejercicio 2: Convertir cadenas en numeros y usar numeros absolutos

```python
planeta1= input("¿Cual es el valor")
planeta1 = int(planeta1)

planeta2= input("¿Cual es el valor")
planeta2 = int(planeta2)

dis_entre_plant_km = abs(planeta2 - planeta1)

dis_entre_plant_mill = abs(dis_entre_plant_km * 0.621)

print('Planeta 1 :',dis_entre_plant_km, 'Kilometros')
print('Planeta 2 :',dis_entre_plant_mill, 'Millas')
```
